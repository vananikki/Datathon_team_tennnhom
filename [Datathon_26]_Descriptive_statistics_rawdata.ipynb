```python
import pandas as pd
```

# 1. Descriptive statistics: "order_items.csv"

```python
df_order_items = pd.read_csv('/content/drive/MyDrive/[2026] Datathon/DATA/order_items.csv')
print(df_order_items.columns)
```

```python
cols_to_stat1 = ['quantity', 'unit_price', 'discount_amount']
stats1 = df_order_items[cols_to_stat1].describe()
print(stats1)
```

```python
print('Mode of columns:')
print(df_order_items[cols_to_stat1].mode())
```

```python
print('Median of columns:')
print(df_order_items[cols_to_stat1].median())
```
